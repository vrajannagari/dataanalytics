```python
def male_female_child(Customer):
    AGE, SEX = Customer
    
    if AGE < 16:
        return 'child'
    else:
        return SEX
```

```python
credit_df['person'] = credit_df[['AGE', 'SEX']].apply(male_female_child, axis=1)
```
